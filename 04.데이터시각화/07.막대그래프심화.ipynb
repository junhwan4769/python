```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)
```

```python
name = df['이름']
height = df['키']
```

```python
plt.barh(name, height, color = 'black')
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160, 210)
plt.show()
```

```python
plt.figure(figsize = (10, 3))
plt.bar(name, height, hatch = '/', color = 'none', edgecolor = 'black' )
plt.show()
```

```python
plt.figure(figsize = (10, 3))
bar = plt.bar(name, height, color = 'none', edgecolor = 'black')
plt.ylim(160,210)
bar[2].set_hatch('/')
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center')
plt.show()
```
