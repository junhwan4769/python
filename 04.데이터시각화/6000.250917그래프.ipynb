```python
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus= False)
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv')
```

```python
df['평균'] = df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
df
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(df['이름'], df['평균'], color = 'black')
ax.set_ylabel('평균점수')
ax.set_ylim(0, 150)
yticks = [y for y in range(0,101,20)]
ax.set_yticks(yticks)


ax2 = ax.twinx()
ax2.set_ylabel('키')
ax2.set_ylim(0, 250)
ax2.plot(df['이름'], df['키'], color='red', marker='o')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(df['이름'], df['평균'], color = 'black')
ax.set_ylabel('평균점수')
ax.set_ylim(0, 150)
yticks = [y for y in range(0,101,20)]
ax.set_yticks(yticks, [f'{y}점' for y in range(0,101,20)])


ax2 = ax.twinx()
ax2.set_ylabel('키')
ax2.set_ylim(0, 250)
ax2.plot(df['이름'], df['키'], color='red', marker='o')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(df['이름'], df['평균'], color = 'black')
ax.set_ylabel('평균점수')
ax.set_ylim(0, 150)
yticks = [y for y in range(0,101,20)]
ax.set_yticks(yticks, [f'{y}점' for y in range(0,101,20)])
for idx, y in enumerate(df['평균']):
    ax.text(idx, y+3, f'{y}점', ha='center')


ax2 = ax.twinx()
ax2.set_ylabel('키')
ax2.set_ylim(0, 250)
ax2.plot(df['이름'], df['키'], color='red', marker='o')
yticks = [y for y in range(0,251,50)]
ax2.set_yticks(yticks, [f'{y}cm' for y in range(0,251,50)])
for idx, y in enumerate(df['키']):
    ax2.text(idx, y+2, f'{y}cm', ha='center')
plt.show()
```
